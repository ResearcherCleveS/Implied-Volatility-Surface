```python
import tkinter as tk
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

class NeuralNetworkGUI:
    def __init__(self, master):
        self.master = master
        master.title("Neural Network GUI")

        # Initialize neural network parameters
        self.input_size = 2
        self.hidden_size = 4
        self.output_size = 1
        self.learning_rate = 0.1
        self.epochs = 100

        # Initialize weights and biases
        self.weights_input_hidden = np.random.rand(self.input_size, self.hidden_size)
        self.bias_hidden = np.zeros((1, self.hidden_size))
        self.weights_hidden_output = np.random.rand(self.hidden_size, self.output_size)
        self.bias_output = np.zeros((1, self.output_size))

        # Create GUI elements
        self.label_lr = tk.Label(master, text="Learning Rate:")
        self.label_lr.grid(row=0, column=0)
        self.entry_lr = tk.Entry(master)
        self.entry_lr.insert(0, str(self.learning_rate))
        self.entry_lr.grid(row=0, column=1)

        # self.label_epochs = "tk.Parameter"
        # self.label_epochs.grid(row=1, column=0)
        self.entry_epochs = tk.Entry(master)
        self.entry_epochs.insert(0, str(self.epochs))
        self.entry_epochs.grid(row=1, column=1)

        self.train_button = tk.Button(master, text="Train", command=self.train)
        self.train_button.grid(row=2, column=0, columnspan=2)

        self.figure, self.ax = plt.subplots()
        self.canvas = FigureCanvasTkAgg(self.figure, master=master)
        self.canvas_widget = self.canvas.get_tk_widget()
        self.canvas_widget.grid(row=3, column=0, columnspan=2)

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
         return x * (1 - x)

    def train(self):
        # Get training parameters from GUI
        self.learning_rate = float(self.entry_lr.get())
        self.epochs = int(self.entry_epochs.get())

        # Simple XOR dataset
        X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
        y = np.array([[0], [1], [1], [0]])

        # Training loop
        for epoch in range(self.epochs):
            # Forward propagation
            hidden_layer_input = np.dot(X, self.weights_input_hidden) + self.bias_hidden
            hidden_layer_output = self.sigmoid(hidden_layer_input)

            output_layer_input = np.dot(hidden_layer_output, self.weights_hidden_output) + self.bias_output
            predicted_output = self.sigmoid(output_layer_input)

            # Backpropagation
            error = y - predicted_output
           
            delta_output = error * self.sigmoid_derivative(predicted_output)
            delta_hidden = np.dot(delta_output, self.weights_hidden_output.T) * self.sigmoid_derivative(hidden_layer_output)

            # Update weights and biases
            self.weights_hidden_output += np.dot(hidden_layer_output.T, delta_output) * self.learning_rate
            self.bias_output += np.sum(delta_output, axis=0, keepdims=True) * self.learning_rate
            self.weights_input_hidden += np.dot(X.T, delta_hidden) * self.learning_rate
            self.bias_hidden += np.sum(delta_hidden, axis=0, keepdims=True) * self.learning_rate

            # Update graph
            self.ax.clear()
            self.ax.plot(error)
            self.ax.set_title('Error During Training')
            self.canvas.draw()

root = tk.Tk()
gui = NeuralNetworkGUI(root)
root.mainloop()
```

```python
import sys
from PyQt5.QtWidgets import (QApplication, QWidget, QLabel, QLineEdit,
                             QPushButton, QVBoxLayout, QHBoxLayout,
                             QMessageBox)
from PyQt5.QtCore import Qt
import numpy as np

class NeuralNetworkGUI(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Neural Network GUI")
        self.initUI()

    def initUI(self):
        # Input fields
        self.layers_label = QLabel("Number of Layers:")
        self.layers_input = QLineEdit()
        self.neurons_label = QLabel("Neurons per Layer:")
        self.neurons_input = QLineEdit()

        # Button
        self.create_button = QPushButton("Create and Train Network")
        self.create_button.clicked.connect(self.create_network)

        # Output label
        self.output_label = QLabel("")
        self.output_label.setAlignment(Qt.AlignCenter)

        # Layout
        input_layout = QHBoxLayout()
        input_layout.addWidget(self.layers_label)
        input_layout.addWidget(self.layers_input)
        input_layout.addWidget(self.neurons_label)
        input_layout.addWidget(self.neurons_input)

        main_layout = QVBoxLayout()
        main_layout.addLayout(input_layout)
        main_layout.addWidget(self.create_button)
        main_layout.addWidget(self.output_label)

        self.setLayout(main_layout)

    def create_network(self):
        try:
            num_layers = int(self.layers_input.text())
            neurons_per_layer = int(self.neurons_input.text())

            # Simple neural network simulation
            input_size = 10
            output_size = 1
            weights = np.random.rand(num_layers, neurons_per_layer, input_size, output_size)
            bias = np.random.rand(num_layers, neurons_per_layer, output_size)
            
            input_data = np.random.rand(input_size)
            
            for i in range(num_layers):
                for j in range(neurons_per_layer):
                    output = np.maximum(0, np.dot(weights[i][j].ravel(), input_data) + bias[i])
                    input_data = output
            
            prediction = np.round(input_data)
            self.output_label.setText(f"Prediction: {prediction}")

        except ValueError:
             QMessageBox.critical(self, "Error", "Please enter valid numbers for layers and neurons.")
#     def create_network(self):
#         num_layers = int(self.layers_input.text())
#         neurons_per_layer = int(self.neurons_input.text())

#         # Simple neural network simulation
#         input_size = 10
#         output_size = 1
#         weights = np.random.rand(num_layers, neurons_per_layer, input_size, output_size)
#         bias = np.random.rand(num_layers, neurons_per_layer, output_size)
            
#         input_data = np.random.rand(input_size)
            
#         for i in range(num_layers):
#             for j in range(neurons_per_layer):
#                 output = np.maximum(0, np.dot(weights[i][j].ravel(), input_data) + bias[i])
#                 input_data = output
            
#             prediction = np.round(input_data)
#             self.output_label.setText(f"Prediction: {prediction}")

        # except ValueError:
        #      QMessageBox.critical(self, "Error", "Please enter valid numbers for layers and neurons.")

if __name__ == '__main__':
    app = QApplication(sys.argv)
    gui = NeuralNetworkGUI()
    gui.show()
    sys.exit(app.exec_())
```
